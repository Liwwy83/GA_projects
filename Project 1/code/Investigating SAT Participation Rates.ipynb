## Background

The SAT and ACT are standardized tests that many colleges and universities in the United States require for their admissions process. This score is used along with other materials such as grade point average (GPA) and essay responses to determine whether or not a potential student will be accepted to the university.

The SAT has two sections of the test: Evidence-Based Reading and Writing and Math ([*source*](https://www.princetonreview.com/college/sat-sections)). The ACT has 4 sections: English, Mathematics, Reading, and Science, with an additional optional writing section ([*source*](https://www.act.org/content/act/en/products-and-services/the-act/scores/understanding-your-scores.html)). They have different score ranges, which you can read more about on their websites or additional outside sources (a quick Google search will help you understand the scores for each test):
* [SAT](https://collegereadiness.collegeboard.org/sat)
* [ACT](https://www.act.org/content/act/en.html)

Standardized tests have long been a controversial topic for students, administrators, and legislators. Since the 1940's, an increasing number of colleges have been using scores from sudents' performances on tests like the SAT and the ACT as a measure for college readiness and aptitude ([*source*](https://www.minotdailynews.com/news/local-news/2017/04/a-brief-history-of-the-sat-and-act/)). Supporters of these tests argue that these scores can be used as an objective measure to determine college admittance. Opponents of these tests claim that these tests are not accurate measures of students potential or ability and serve as an inequitable barrier to entry. Lately, more and more schools are opting to drop the SAT/ACT requirement for their Fall 2021 applications ([*read more about this here*](https://www.cnn.com/2020/04/14/us/coronavirus-colleges-sat-act-test-trnd/index.html)).

1. Manually calculate mean:

    Write a function that takes in values and returns the mean of the values. Create a list of numbers that you test on your function to check to make sure your function works!

```python
# Code:
def mean(num_list):
    """Takes in a list of numbers and returns its arithmetic mean."""
    sum = 0
    for num in num_list:
        sum += num
    return sum / len(num_list)

num_list = [1, 4, -4, 0, 30, 11]
mean(num_list)
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Code:
def std_dev(num_list):
    """Takes in a list of numbers and returns its standard deviation."""
    mu = mean(num_list)
    var = mean([(num - mu) ** 2 for num in num_list])
    return var ** 0.5
        
print(std_dev(num_list))

def unbiased_std_dev(num_list):
    """Takes in a list of numbers and returns its UNBIASED ESTIMATE for population standard deviation."""
    """This is done by dividing by (n - 1) instead of n."""
    n = len(num_list)
    mu = mean(num_list)
    var = mean([(num - mu) ** 2 for num in num_list]) * n / (n - 1)
    return var ** 0.5
        
print(unbiased_std_dev(num_list))
```

### Standard deviations
Standard deviations provided by pandas describe method are all higher than the ones calculated using our own function.
This is because the describe method treats the data as a sample and calculates the unbiased estimate for population standard deviation, adjusting for the degree of freedom by dividing by (sample size - 1) instead of (sample size) in the formula. 

We verify this below by creating another function called unbiased_std_dev and checking that the values tally.

```python
# Calculate unbiased estimates of standard deviations for all features in ACT and SAT merged dataframes
act_unbiased_std = {x: round(unbiased_std_dev(act_merged[x]), 3) for x in act_merged.set_index('state')}
sat_unbiased_std = {x: round(unbiased_std_dev(sat_merged[x]), 3) for x in sat_merged.set_index('state')}

# Calculate standard deviations for features in ACT 2017 excluding 'National'
act_2017_unbiased_std = {x: round(unbiased_std_dev(act_2017[1:][x]), 3) for x in act_2017[1:].set_index('state')}
```

## Conclusions and Recommendations

### Factors influencing SAT participation rates
Higher SAT participation rates are strongly associated with higher SAT participation rates in other years, lower ACT participation rates, and lower total scores.

#### SAT participation rates in other years
We are unable to change historical participation rates, thus we should focus on the other 2 factors.

#### Total scores
It is highly possible that the strong negative correlation is because in states with low participation rates, only the higher ability students took the test. Thus even if we intentionally deflate SAT total scores, this is unlikely to lead to an increase in participation rates.

#### ACT participation rates
We can work on trying to convince more states to switch over from ACT (or other tests) to SAT.
In particular, we focus on the three states of Colorado, Illinois and West Virginia, which had drastic increase in their SAT participation rates from less than 20% to 100% or almost 100%. We research into these 3 states to see what happened to cause this change.

### Research into specific states
**Colorado**
* [Colorado's switch from ACT to SAT](https://www.coloradokids.org/wp-content/uploads/2016/01/ACTvsSAT_FINAL.pdf)
Since 2001, Colorado has required every high school junior to take the ACT as their 11th grade college entrance
exam, for which the state covered the cost. In December of 2015, a selection committee voted to adopt SAT as the required test instead.

Among the reasons quoted are: 
1. The SAT is aligned to the Common Core standards in English and math, which are a part of the Colorado Academic Standards. The redesigned SAT measures the skills and knowledge that evidence shows is critical for college and career readiness, including critical thinking and analytical reasoning. 
2. In March 2015, College Board partnered with Khan Academy to provide free, personalized test prep for any student with computer access, leveling the playing field for families that might not be able to afford test prep or tutors for their children.

**Illinois**
* [Illinois moves ahead with new testing plan, replacing ACT with SAT](https://www.chicagotribune.com/news/ct-illinois-chooses-sat-met-20160211-story.html)
In 2016, College Board won a three-year, $14.3 million bid to give its exam to all public high school juniors in Illinois.

**West Virginia**
* [WV chooses SAT as new high school standardized test for juniors](https://www.wvgazettemail.com/news/education/wv-chooses-sat-as-new-high-school-standardized-test-for-juniors/article_b60d2618-4943-56f6-b180-4b4442172ef8.html)
The West Virginia Department of Education has chosen SAT as the new statewide standardized test to replace the Smarter Balanced exam for high school students in spring 2018.

### Recommendations
1. Continuously monitor any changes to the Common Core standards and align the SAT to them.
2. Actively reach out to the Departments of Education in more states to persuade them of the strengths of the SAT and the benefits it offers to their students.
3. Continue partnering with Khan Academy to provide free test prep. Review test prep resources to ensure quality and alignment to the SAT.
